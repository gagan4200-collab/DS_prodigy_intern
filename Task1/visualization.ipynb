### Importing the necessary python libraries and the data set.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population_data = pd.read_csv('population.csv')
```

### Extracting the relevent columns

```python
data_2023 = population_data.loc[:, ['Country Name', '2023']]
```

### Remove rows with missing values

```python
data_2023.dropna()
```

### Display the first few rows of the cleaned data

```python
data_2023.head()
```

### We will then visualize the data using a Histogram.

```python

sns.set(style="whitegrid")

# Histogram for population distribution in 2023
plt.figure(figsize = (12, 6))
sns.histplot(data = data_2023, x='2023', bins=30, kde=True, color='blue')
plt.title('Population Distribution in 2023')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

### We will then visualize the data using a Bar Chart.

```python

data_2023['2023'] = pd.to_numeric(data_2023['2023'], errors='coerce')

# Sort data by population in descending order and select top 10
top_10_countries = data_2023.sort_values(by='2023', ascending=False).head(10)

# Set up the visual style
sns.set(style="whitegrid")

# Bar chart for top 10 most populous countries in 2023
plt.figure(figsize=(12, 6))
sns.barplot(data = top_10_countries, x='Country Name', y='2023', palette='viridis')
plt.title('Top 10 Most Populous Countries in 2023')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
